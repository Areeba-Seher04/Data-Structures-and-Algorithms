## Telephone Directory of total Employees in Company.

```python
telephone_directory= pd.read_csv("Telephone_Directory_of_Total_Employees_in_company.csv")
telephone_directory  #csv file of total employees
```

### m=97 2 bytes (length of hashtable = 00-99 )

```python
hashtable = [[0,0,0,0] for i in range(100)]
```

### Hashing_function calculates the hash key of employee number

```python
def hashing_func(key):
  location = key%97
  return location

```

## insert(hashtable,np.array[employee numbers]) 
#### you can pass list of many/one employee numbers.

```python
def insert(hashtable,s):
    key = s %97
    print("Hashed keys: ",key)    ## print hashing key of all Employee numbers
    i = 0
    
    while i < len(key):
        h=telephone_directory[telephone_directory.Employee_Number==s[i]]
        if hashtable[key[i]][1]==s[i]:
            print("Employee number already exists") #if same employee number passed in function.
            i=i+1
        elif hashtable[key[i]][0]==0:
             hashtable[key[i]][0]=h.iloc[0,0]
             hashtable[key[i]][1]=s[i]
             hashtable[key[i]][2]=h.iloc[0,2]
             hashtable[key[i]][3]=h.iloc[0,3]
             i = i+1
        else:
             j=1
             q=1
             while q<97:
                 keys=(key[i]+j)%97
                 if hashtable[keys][0]==0:
                     hashtable[keys][0]=h.iloc[0,0]
                     hashtable[keys][1]=s[i]
                     hashtable[keys][2]=h.iloc[0,2]
                     hashtable[keys][3]=h.iloc[0,3]
                     i = i+1
                     break
                 else:
                    j=(q+1)**2
                    q=q+1
                

l = list(telephone_directory['Employee_Number'][0:25])  # l is a list of employee numbers
print("Employee number:",l)           # print all Employee numbers passed by user as input 
s=np.array(l)                         #s is an array of employee numbers
insert(hashtable,s)                   
insert(hashtable,np.array([45206,10101,44585,80491,96617]))
hashtable2 = pd.DataFrame(hashtable,columns=['Name','Employee_Number','Landline_number','Mobile_Number'])
hashtable2 
```

### search(hashtable,Employee_number)
#### returns all information of Employee from hashtable as well as on which index it is inserted in hashtable.

```python
def search(hashtable,key):
    loc = hashing_func(key)
    if key == hashtable2.iloc[loc][1]:
        print(hashtable2.iloc[loc])
    else:
        i=1
        q = 1
        while q<97:
             locs=(loc+i)%97
             if key==hashtable2.iloc[locs][1]:
                 print(hashtable2.iloc[locs])
                 break
             else:
                i=(q+1)**2
                q = q+1
        
       
search(hashtable,45206)
search(hashtable,10101)
search(hashtable,83537)
search(hashtable,59152)
search(hashtable,80491)
```

### Probes(Employee number in list format)
#### Probes function can calculate successful probes and unsuccessful probes. Employee number in list should be 1 at a time.

```python
def Probes(l):
     i = 0
     j = 0
     while j<len(l):
        
        if l[0] == hashtable2.iloc[hashing_func(l[j])][1]:
            i = i+1
            j = j+1
    
            print("successful Probes: ",i)
            break
        else:
            m = 0
            q = 0
            while l[j] !=hashtable2.iloc[hashing_func(l[j])][1]:
                k = (hashing_func(l[j])+m)%97
                
                if l[j] == hashtable2.iloc[k][1]:
                    i = i+1
                    j = j+1
                    print("Successful Probes: ",i)
                    break
                elif hashtable2.iloc[k][1]==0:
                    i = i+1
                    print("Unsuccesful Probes: ",i)
                    break
                else:
                    
                    m = (q+1)**2
                    q = q+1 
                    i=i+1
                
        break 
    
    
                
                     
Probes([37405])
Probes([16453])
Probes([78048])
Probes([83537])
Probes([45206])
Probes([37406])
Probes([12360])
Probes([56320])
Probes([56326])
Probes([10034])
```
